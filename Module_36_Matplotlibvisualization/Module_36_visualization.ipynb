Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Ans

Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide range of functionalities for creating various types of plots and charts.

Matplotlib is used for:

Data visualization: Matplotlib allows users to create a wide variety of plots and charts to visualize data in a clear and understandable manner. Exploration and analysis: It helps in exploring and analyzing data by visualizing trends, patterns, and relationships within the data. Presentation and communication: Matplotlib enables users to create high-quality visualizations for presentations, reports, and publications. Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line plot: Used to visualize the relationship between two variables by connecting data points with straight lines. Scatter plot: Used to display the relationship between two variables by plotting individual data points on a two-dimensional plane. Histogram: Used to represent the distribution of a continuous variable by dividing the data into bins and displaying the frequency of observations in each bin. Bar plot: Used to compare the values of different categories or groups by representing them as rectangular bars with heights proportional to the values they represent. Pie chart: Used to represent the distribution of a categorical variable by dividing a circle into sectors, where each sector represents a category and its area or angle represents the proportion of that category in the data.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data

plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plot scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.7)  # alpha controls the transparency of points
plt.title('Scatter Plot of X and Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)  # Add grid lines
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title('Line Plot 4')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit ($)')
plt.title('Bar Plot of Company Profit')
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='salmon')
plt.xlabel('Profit ($)')
plt.ylabel('Company')
plt.title('Horizontal Bar Plot of Company Profit')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine data into a list
data = [box1, box2]

# Box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data)
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Two Datasets')
plt.grid(True)
plt.show()
```
