Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans -1

Scatter Plot:

Use: A scatter plot is used to visualize the relationship between two continuous variables. It helps in identifying patterns, clusters, and correlations in the data.
Histogram:

Use: Histograms are used to display the distribution of a single continuous variable. They provide insights into the central tendency, spread, and shape of the data distribution.
Bar Plot:

Use: Bar plots are used to compare categorical data across different groups or categories. They are useful for visualizing counts, frequencies, or aggregated values of categorical variables.
Box Plot:

Use: Box plots, also known as box-and-whisker plots, are used to visualize the distribution of a continuous variable across different categories or groups. They display the median, quartiles, and potential outliers in the data.
Heatmap:

Use: Heatmaps are used to visualize the matrix-like data where each cell represents the value of the corresponding row and column. They are commonly used in correlation matrices, confusion matrices, and categorical data analysis.

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Plot two box plots
plt.figure(figsize=(10, 5))

# Box plot for age
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

# Box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

# Display the plots
plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot the histogram with hue for the 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack', bins=30)
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title='Cut')
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Plot the pair plot with hue for the 'species' column
sns.pairplot(data=iris_data, hue='species')
plt.title('Pair Plot of Iris Dataset')
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the dataset
flights_pivot = flights_data.pivot("month", "year", "passengers")

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data=flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.title('Passenger Count by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
